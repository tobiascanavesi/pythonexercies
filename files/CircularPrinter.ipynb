```python
graph2={"A" : "B", "B" : "C", "C" : "D", "D" : "E", "E" : "F", "F" : "G", 
 "G" : "H", "H" : "I", "I" : "J", "J" : "K", "K" : "L", "L" : "M", 
 "M" : "N", "N" : "O", "O" : "P", "P" : "Q", "Q" : "R", "R" : "S", 
 "S" : "T", "T" : "U", "U" : "V", "V" : "W", "W" : "X", "X" : "Y", 
 "Y" : "Z", "Z" : "A"}
```

```python
def find_shortest_path(graph, start, end, path=[]):
    path = path + [start]
    if start == end:
        return path
    if start not in graph:
        return None
    shortest = None
    for node in graph[start]:
        if node not in path:
            newpath = find_shortest_path(graph, node, end, path)
            if newpath:
                if not shortest or len(newpath) < len(shortest):
                    shortest = newpath
    return shortest
```

```python
def getTime(s):
    lista=[]
    s='A'+s
    for i in range(len(s)-1):
        lista.append(min(len(find_shortest_path(graph2, s[i], s[i+1])),len(find_shortest_path(graph2, s[i+1], s[i])))-1)
    return sum(lista)
```
